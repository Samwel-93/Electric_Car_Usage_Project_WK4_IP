```python
#Load the exported dataframe
e_car2 = pd.read_csv('electric_cars_usage.csv')
e_car2.head()
```

```python
# we begin by creating another copy of our dataset
#
e_car3 = e_car2
e_car3.head(5)
```

```python
# create a date column by combining year, month, and day
#
e_car3['Date'] = e_car3["year"].map(str) + ':' + e_car3["month"].map(str) + ':' + e_car3["day"].map(str)
e_car3.head(5)
```

```python
# Create a time column by combining hour and minute columns
#
e_car3['Time'] = e_car3["hour"].map(str) + ':' + e_car3["minute"].map(str)
e_car3.head(5)
```

```python
# Add a seconds column to inorder to have a complete time 
#
secs = [0]*5000
e_car3['Seconds'] = secs
e_car3.head(4)
```

```python
# Create a complete date_time column by combining Date and Time columns
#
e_car3['Date_Time'] = e_car3["Date"].map(str) + ':' + e_car3["Time"].map(str)+ ':' + e_car3["Seconds"].map(str)
e_car3.head(5)
```

```python
# let us now change date_time data type to datetime format
from datetime import datetime
e_car3['Date_Time'] = pd.to_datetime(e_car3['Date_Time'] , format='%Y:%m:%d:%H:%M:%S', errors = 'coerce')
e_car3.head(5)
```

```python
# Let us drop year, month, day, minute date, time, and seconds columns
#
e_car4 = e_car3.drop(['year', 'month', 'day', 'minute', 'Date','Time','Seconds'] , axis=1)
e_car4.head(5)
```

## 1.5 Answering the Research Question
Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

**HINT:**

To compute usage, 

we will need to understand that we have to join successive (in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.

```python
# electric car services in Paris only
#
e_car_paris = e_car4[e_car4['City'] =='Paris'] 
```

```python
#print entries in Paris only
e_car_paris.head(5)
```

```python
# group dataset by ID, date_time and hour
#
e_car_paris2 = e_car_paris.groupby(['ID','Date_Time', 'hour']).sum()
e_car_paris2.head(10)
```

```python
# Summary of e_car_paris2 dataframe
e_car_paris2.describe(include='all')
```

```python
# We find the max absolute difference 
#
e_car_paris2['Bluecar_diff'] = e_car_paris2['Bluecar counter'].diff()
e_car_paris2.head(5)
```

```python
#Let us sort Bluecar_diff column to obtain positive values. 
# Positive values indicate that the car was picked
#
e_car_paris2 = e_car_paris2[e_car_paris2['Bluecar_diff'] > 0]
e_car_paris2.head(5)
```

```python
# Let us sort Bluecar_diff column to obtain positive values. 
e_car_paris2.groupby(['hour'])['Bluecar_diff'].sum().sort_values(ascending  = False)
```
